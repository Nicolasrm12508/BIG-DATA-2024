```python
import os
import supervision as sv
from groundingdino.util.inference import load_image, predict

# Define las rutas de las carpetas
input_folder = "/content/Imagenes"
output_folder = "/content/Resultados"

# Crea la carpeta de salida si no existe
os.makedirs(output_folder, exist_ok=True)

# Carga el modelo GroundingDINO (asegúrate de haberlo cargado previamente)
# model = ... # Asegúrate de cargar tu modelo aquí

# Parámetros de predicción
BOX_THRESHOLD = 0.26
TEXT_THRESHOLD = 0.50
text = "car plate"

# Itera sobre todas las imágenes en la carpeta de entrada
for filename in os.listdir(input_folder):
    if filename.endswith(".png") or filename.endswith(".jpg"):  # Filtra por extensiones de imagen
        image_path = os.path.join(input_folder, filename)

        # Carga la imagen
        image_source, image = load_image(image_path)

        # Realiza la predicción
        boxes, logits, phrases = predict(
            model=model,
            image=image,
            caption=text,
            box_threshold=BOX_THRESHOLD,
            text_threshold=TEXT_THRESHOLD,
            device='cpu'  # Cambiado a 'cpu'
        )

        # Genera el archivo de etiquetas en formato YOLO
        label_file_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.txt")

        with open(label_file_path, 'w') as label_file:
            for box, phrase in zip(boxes, phrases):
                # Extrae las coordenadas del cuadro delimitador (normalizadas)
                x_center = (box[0] + box[2]) / 2 / image.shape[1]
                y_center = (box[1] + box[3]) / 2 / image.shape[0]
                width = (box[2] - box[0]) / image.shape[1]
                height = (box[3] - box[1]) / image.shape[0]

                # Asumiendo que la etiqueta es un número; puedes mapear frases a índices si es necesario
                label_index = 0  # Cambia esto según tu mapeo de etiquetas

                # Escribe en el archivo en formato YOLO: <label> <x_center> <y_center> <width> <height>
                label_file.write(f"{label_index} {x_center} {y_center} {width} {height}\n")
```

```python
import supervision as sv
from groundingdino.util.inference import load_image, predict, annotate
from segment_anything import SamPredictor, sam_model_registry
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --- Configuración del modelo GroundingDINO ---
# Define los parámetros del modelo
image_path = "/content/Imagenes/Cars1.png"
text = "car plate"  # Texto para buscar objetos
BOX_THRESHOLD = 0.26
TEXT_THRESHOLD = 0.50

# Cargar la imagen
image_source, image = load_image(image_path)

# Cargar el modelo GroundingDINO (asegúrate de que esté configurado)
from groundingdino.util.inference import load_model, load_image, predict, annotate

model = load_model("/content/GroundingDINO/groundingdino/config/GroundingDINO_SwinT_OGC.py",
     "weights/groundingdino_swint_ogc.pth")
# Ejecutar predicción con GroundingDINO
boxes, logits, phrases = predict(
    model=model,
    image=image,
    caption=text,
    box_threshold=BOX_THRESHOLD,
    text_threshold=TEXT_THRESHOLD,
    device="cpu"
)

# Visualizar las cajas detectadas
annotated_frame = annotate(
    image_source=image_source,
    boxes=boxes,
    logits=logits,
    phrases=phrases
)

# Mostrar imagen con cajas de detección
plt.imshow(annotated_frame)
plt.title("Detecciones con GroundingDINO")
plt.axis("off")
plt.show()

# --- Configuración del modelo SAM ---
# Cargar el modelo SAM
sam_checkpoint = "sam_vit_h_4b8939.pth"  # Ruta al checkpoint de SAM
sam = sam_model_registry["vit_h"](checkpoint=sam_checkpoint)
predictor = SamPredictor(sam)

# Convertir la imagen al formato necesario para SAM
image_bgr = cv2.cvtColor(image_source, cv2.COLOR_RGB2BGR)
predictor.set_image(image_bgr)

# Convertir las cajas detectadas por GroundingDINO al formato de SAM
# Conversión manual de formato [x_min, y_min, x_max, y_max] a [x, y, width, height]
boxes_xywh = []
for box in boxes:
    x_min, y_min, x_max, y_max = box
    width = x_max - x_min
    height = y_max - y_min
    boxes_xywh.append([x_min, y_min, width, height])
boxes_xywh = np.array(boxes_xywh)

# Generar máscaras de segmentación para cada caja detectada
masks = []
for box in boxes_xywh:
    mask, _, _ = predictor.predict(box=box, multimask_output=False)
    masks.append(mask)

# --- Mostrar las máscaras generadas ---
for i, mask in enumerate(masks):
    plt.figure()
    plt.imshow(mask[0], cmap="gray")
    plt.title(f"Máscara {i+1}")
    plt.axis("off")
    plt.show()
```
